```r
# Install and load the googledrive package
install.packages("googledrive")
library(googledrive)

# Authenticate Google Drive access
drive_auth()

# List files in your Google Drive
drive_find(n_max = 10)
```

```r
# Find the file by name and get its ID
file <- drive_get("WorldCupMatches.csv")

# Download the file to the Colab environment
drive_download(file, path = "WorldCupMatches.csv", overwrite = TRUE)

# Read the CSV file
world_cup <- read.csv("WorldCupMatches.csv", stringsAsFactors=TRUE)

# View first few rows to confirm it loaded correctly
head(world_cup)


```

```r
# Ensure dplyr is installed and loaded
install.packages("dplyr")
library(dplyr)

# Total goals per home team
goals_per_team <- world_cup %>%
  group_by(Home.Team.Name) %>%
  summarise(total_goals = sum(Home.Team.Goals, na.rm=TRUE))

cat("\n(e) Total Goals by Home Team:\n")
print(head(goals_per_team, 10))  # Show only first 10 for readability

# Average attendance per year
avg_attendance_per_year <- world_cup %>%
  group_by(Year) %>%
  summarise(avg_attendance = mean(Attendance, na.rm=TRUE))

cat("\n(f) Average Attendance Per Year (Top 10 shown):\n")
print(head(avg_attendance_per_year, 10))
```
