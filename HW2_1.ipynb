```r
# ======================================================
# Sales Data Analysis & Visualization
# ======================================================

# Load required libraries
library(dplyr)
library(ggplot2)

# Create the dataframe
df1 <- data.frame(

  Name = c('James', 'Paul', 'Richards', 'Marico', 'Samantha', 'Ravi', 'Raghu',
           'Richards', 'George', 'Ema', 'Samantha', 'Catherine'),

  State = c('Alaska', 'California', 'Texas', 'North Carolina', 'California', 'Texas',
            'Alaska', 'Texas', 'North Carolina', 'Alaska', 'California', 'Texas'),

  Sales = c(14, 24, 31, 12, 13, 7, 9, 31, 18, 16, 18, 14)
)

# Summarize total sales by state
sales_summary <- df1 %>%
  group_by(State) %>%
  summarise(Total_Sales = sum(Sales)) %>%
  arrange(desc(Total_Sales))  # Sort in descending order

# Print the summarized sales table
print(sales_summary)

# Summary statistics of sales data
cat("\nSummary Statistics of Sales Data:\n")
print(summary(df1$Sales))

# Bar plot with data labels
ggplot(sales_summary, aes(x = reorder(State, -Total_Sales), y = Total_Sales, fill = State)) +

  geom_bar(stat = "identity") +

  geom_text(aes(label = Total_Sales),
            vjust = -0.5, size = 5, fontface = "bold") +  # Add data labels

  labs(title = "Total Sales by State",
       x = "State",
       y = "Total Sales") +

  theme_minimal() +

  theme(legend.position = "none")

# End of Script

```

```r
# Ensure required libraries are installed and loaded
install.packages("dplyr")
install.packages("ggplot2")
install.packages("googledrive")
library(dplyr)
library(ggplot2)
library(googledrive)

# Authenticate Google Drive access
drive_auth()

# Define file path from Google Drive (replace FILE_NAME with actual file name)
file_name <- "World_Cup_Matches.csv"
drive_download(file_name, path = file_name, overwrite = TRUE)

# Load dataset
world_cup <- read.csv(file_name)

# Verify dataset is loaded correctly
if (!"world_cup" %in% ls()) {
  stop("Error: Dataset 'world_cup' not found. Please check the file path and reload.")
}

# Check column names
required_columns <- c("Year", "Attendance")
missing_columns <- setdiff(required_columns, colnames(world_cup))
if (length(missing_columns) > 0) {
  stop(paste("Error: Missing required columns:", paste(missing_columns, collapse=", "))) }

# Average attendance per year
avg_attendance_per_year <- world_cup %>%
  group_by(Year) %>%
  summarise(avg_attendance = mean(Attendance, na.rm = TRUE))

cat("\n(f) Average Attendance Per Year (Top 10 shown):\n")
print(head(avg_attendance_per_year, 10))

# Plot attendance trends over time with improvements
ggplot(avg_attendance_per_year, aes(x = Year, y = avg_attendance)) +
  geom_line(color = "blue", linewidth = 1) +  # Fix deprecated 'size' warning
  geom_point(color = "red", size = 2) +  # Add points for better visibility
  geom_text(aes(label = round(avg_attendance, 0)), vjust = -0.5, size = 3.5) +  # Add data labels
  geom_smooth(method = "loess", color = "black", linetype = "dashed") +  # Add trend line
  labs(title = "Enhanced Average Attendance Per Year",
       x = "Year",
       y = "Average Attendance") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))  # Rotate x-axis labels
```

```r
# ======================================================
# Ensure required libraries are installed and loaded
# ======================================================

install.packages("dplyr")  # Install if not installed
library(dplyr)

# ======================================================
# Load the dataset
# ======================================================

file_path <- "metabolites.csv"  # File downloaded from Google Drive
metabolites <- read.csv(file_path, stringsAsFactors = TRUE)

# Display first few rows to verify data
head(metabolites)

# ======================================================
# (a) Count Alzheimer's Patients
# ======================================================

alzheimers_count <- sum(metabolites$Label == "Alzheimer", na.rm = TRUE)

cat("\n(a) Total number of Alzheimer's patients:", alzheimers_count, "\n")

# ======================================================
# (b) Count Missing Values per Column
# ======================================================

missing_values <- colSums(is.na(metabolites))

cat("\n(b) Missing values per column (only columns with missing values are shown):\n")
print(missing_values[missing_values > 0])  # Show only columns with missing data

# ======================================================
# (c) Remove Rows with Missing 'Dopamine' Values
# ======================================================

clean_metabolites <- metabolites %>%
    filter(!is.na(Dopamine))

cat("\n(c) Data shape after removing rows with missing Dopamine values:",
    dim(clean_metabolites)[1], "rows and", dim(clean_metabolites)[2], "columns\n")

# ======================================================
# (d) Replace Missing 'c4-OH-Pro' Values with Median
# ======================================================

median_c4OHPro <- median(clean_metabolites$c4.OH.Pro, na.rm = TRUE)

clean_metabolites <- clean_metabolites %>%
    mutate(c4.OH.Pro = ifelse(is.na(c4.OH.Pro), median_c4OHPro, c4.OH.Pro))

cat("\n(d) Missing values in 'c4-OH-Pro' replaced with median:", median_c4OHPro, "\n")

# ======================================================
# (e) Drop Columns with >25% Missing Values
# ======================================================

threshold <- 0.25 * nrow(clean_metabolites)  # Define threshold for missing values

columns_dropped <- names(clean_metabolites)[colSums(is.na(clean_metabolites)) >= threshold]

clean_metabolites <- clean_metabolites %>%
    select(which(colSums(is.na(.)) < threshold))

cat("\n(e) Columns dropped due to more than 25% missing values:",
    ifelse(length(columns_dropped) > 0, paste(columns_dropped, collapse=", "), "None"), "\n")

cat("\nUpdated dataset shape:", dim(clean_metabolites)[1], "rows and", dim(clean_metabolites)[2], "columns\n")

# ======================================================
# Display Summary of Cleaned Data
# ======================================================

cat("\nFinal Cleaned Data Summary:\n")
summary(clean_metabolites)
```
