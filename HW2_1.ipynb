```r
# Ensure dplyr is installed and loaded
install.packages("dplyr")
library(dplyr)

# Total goals per home team
goals_per_team <- world_cup %>%
  group_by(Home.Team.Name) %>%
  summarise(total_goals = sum(Home.Team.Goals, na.rm=TRUE))

cat("\n(e) Total Goals by Home Team:\n")
print(head(goals_per_team, 10))  # Show only first 10 for readability

# Average attendance per year
avg_attendance_per_year <- world_cup %>%
  group_by(Year) %>%
  summarise(avg_attendance = mean(Attendance, na.rm=TRUE))

cat("\n(f) Average Attendance Per Year (Top 10 shown):\n")
print(head(avg_attendance_per_year, 10))

# Load ggplot2 library
library(ggplot2)

# Plot attendance trends over time with improvements
ggplot(avg_attendance_per_year, aes(x = Year, y = avg_attendance)) +
  geom_line(color = "blue", linewidth = 1) +  # Fix deprecated 'size' warning
  geom_point(color = "red") +
  geom_smooth(method = "loess", color = "black", linetype = "dashed") +  # Add trend line
  labs(title = "Average Attendance Per Year",
       x = "Year",
       y = "Average Attendance") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))  # Rotate x-axis labels
```

```r
# ===============================
# 📌 Ensure required libraries are installed and loaded
# ===============================
install.packages("dplyr")  # Install if not installed
library(dplyr)

# ===============================
# 📌 Load the dataset
# ===============================
file_path <- "metabolites.csv"  # File downloaded from Google Drive
metabolites <- read.csv(file_path, stringsAsFactors=TRUE)

# ✅ Display first few rows to verify data
head(metabolites)

# ===============================
# 📌 (a) Count Alzheimer's Patients
# ===============================
alzheimers_count <- sum(metabolites$Label == "Alzheimer", na.rm=TRUE)

cat("\n(a) Number of Alzheimer's patients:", alzheimers_count, "\n")

# ===============================
# 📌 (b) Count Missing Values per Column
# ===============================
missing_values <- colSums(is.na(metabolites))

cat("\n(b) Missing values per column:\n")
print(missing_values)

# ===============================
# 📌 (c) Remove Rows with Missing 'Dopamine' Values
# ===============================
clean_metabolites <- metabolites %>%
    filter(!is.na(Dopamine))

cat("\n(c) DataFrame after removing missing Dopamine values:", dim(clean_metabolites), "\n")

# ===============================
# 📌 (d) Replace Missing 'c4-OH-Pro' Values with Median
# ===============================
median_c4OHPro <- median(clean_metabolites$c4.OH.Pro, na.rm=TRUE)

clean_metabolites <- clean_metabolites %>%
    mutate(c4.OH.Pro = ifelse(is.na(c4.OH.Pro), median_c4OHPro, c4.OH.Pro))

cat("\n(d) Missing values in 'c4-OH-Pro' replaced with median value:", median_c4OHPro, "\n")

# ===============================
# 📌 (e) (Optional) Drop Columns with >25% Missing Values
# ===============================
threshold <- 0.25 * nrow(clean_metabolites)  # Define threshold

clean_metabolites <- clean_metabolites %>%
    select(which(colSums(is.na(.)) < threshold))

cat("\n(e) Columns with more than 25% missing values removed. Updated shape:", dim(clean_metabolites), "\n")

# ===============================
# 📌 Display Summary of Cleaned Data
# ===============================
cat("\n✅ Final Cleaned Data Summary:\n")
summary(clean_metabolites)
```
